```python
from rep.report import metrics as rep_metrics
```

```python
report.compute_metric(rep_metrics.RocAuc())
```

```python
learning_curve = report.learning_curve(rep_metrics.RocAuc(), metric_label='ROC AUC', steps=1)
learning_curve.plot(new_plot=True)
```
